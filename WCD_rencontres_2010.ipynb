```python
import lxml
import lxml.html
from time import sleep
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import csv
import requests
import numpy as np
import pandas as pd
import re
import random
from random import randint
```

```python
url = "https://www.lequipe.fr/Football/Coupe-du-monde/Saison-2010/calendrier-resultats.html"
#url="https://www.lequipe.fr/Football/Coupe-du-monde/Saison-2014/calendrier-resultats.html"
#url="https://www.lequipe.fr/Football/Coupe-du-monde/Saison-2018/calendrier-resultats.html"
r = requests.get(url)
#print(r.status_code)
soup = bs(r.content)
```

```python
df_match_2010 =  pd.DataFrame(columns=['phase','nom_pays1', 'score','nom_pays2'])
```

```python
df_match_2010
```

```python
section = soup.find('section',{'class': 'tab-date'})
```

```python
t = section.find_all("tr")
```

```python
for elem in t:
    print(elem.text)
```

```python
for row in t:
    matchinfos= []
    element = row.find_all("td")
    
    for data in element:
        matchinfos.append(data.text)
        try:
            df_match_2010 = df_match_2010.append({'phase':matchinfos[0],'nom_pays1':matchinfos[2],'score':matchinfos[4],'nom_pays2':matchinfos[6]}, ignore_index=True) 
        except IndexError:
            pass
        print(df_match_2010)
```

```python
df_match_2010 = df_match_2010.drop_duplicates(keep='first')
```

```python
df_match_2010
```

```python
df_match_2010.to_csv('match_2010.csv', sep='\t', encoding='utf-8')
```
