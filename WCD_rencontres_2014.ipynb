```python
import lxml
import lxml.html
from time import sleep
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import csv
import requests
import numpy as np
import pandas as pd
import re
import random
from random import randint
```

```python
url="https://www.lequipe.fr/Football/Coupe-du-monde/Saison-2014/calendrier-resultats.html"
r = requests.get(url)
#print(r.status_code)
soup = bs(r.content)
```

```python
df_match_2014=pd.DataFrame(columns=['phase','nom_pays1', 'score','nom_pays2'])
```

```python
#wrapper = soup.find('div', {'class': 'bloc-liste'}) #get informations
#rows = wrapper.find_all('section', {'class' : 'mainDate'}) #store teams name
```

```python
section = soup.find('section',{'class': 'tab-date'})
```

```python
t = section.find_all("tr")
#t = section.find_all("tr", {'class' : re.compile("^alternate")})
```

```python
for row in t:
    matchinfos= []
    element = row.find_all("td")
    
    for data in element:
        matchinfos.append(data.text.strip())
        try:
            df_match_2014 = df_match_2014.append({'phase':matchinfos[0],'nom_pays1':matchinfos[1],'score':matchinfos[3],'nom_pays2':matchinfos[5]}, ignore_index=True) 
        except IndexError:
            pass
        print(df_match_2014)
    
```
