# WebScraping - Site : L'Equipe - buts des joueurs - cdm 2018

```python
import lxml
import lxml.html
from time import sleep
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import csv
import requests
import numpy as np
import pandas as pd
import re
import random
from random import randint
```

```python
url="https://www.lequipe.fr/Football/FootballClassementChampionnat2768_BUT_1.html"
r = requests.get(url)
#print(r.status_code)
soup = bs(r.content)
```

```python
#Create DF
df_goals = pd.DataFrame(columns=['annee','nom_joueur', 'nb_buts', 'nom_pays'])
```

```python
df_goals
```

```python
#Find the container with all teams informations
wrapper = soup.find('tbody')
```

```python
table = wrapper.find_all("tr", {'class': 'js-tr-hover'})
```

```python
for r in table:
    td = r.find("span", {'class' : 'in2'})
    td=td.text
    country = r.find('div', {'class' :'countryName'})
    country=country.text
    goal= r.find("td", {'class' : 'stat stat-large highlighted'})
    goal=goal.text
    df_goals = df_goals.append({'nom_joueur':td, 'nb_buts':goal, 'nom_pays':country}, ignore_index=True)
```

```python
df_goals.to_csv('goals_2018.csv', sep='\t', encoding='utf-8')
```
